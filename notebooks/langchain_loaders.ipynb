```python
def process_pdf(pdf):
    from langchain.document_loaders import PyPDFLoader
    loader = PyPDFLoader(pdf)
    pages = loader.load()
    text = ''
    for page in pages:
        text += page.page_content
    return text
```

```python
pdf = '../raw_data/sample.pdf'
print(process_pdf(pdf))
```
